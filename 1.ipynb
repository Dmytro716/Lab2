```python
from random import randint

for i in range(3):
    dice = "Lucky" if randint(1, 6) == 4 else "Unlucky"
    print(f"Нам випало: {dice}")
    if dice == "Lucky":
        print("Ми виграли, нам випало Lucky!")
        break
    print("Хай пощастить наступного разу!")
```

```python
i = -1
dice = None

while i <= 3 and dice != "Lucky":  # умова виконується на початку ітерації
    dice = "Lucky" if randint(1, 6) == 4 else "Unlucky"
    print(f"Нам випало: {dice}")
    i += 1

if dice == "Lucky":
    print(f"Ми виграли, нам випало {dice}!")
else:
    print("Хай пощастить наступного разу!")
```

```python
mmd = ["Червоний", "Жовтий", "Зелений", "Синій", "Коричневий", "Оранжевий"]
factory = mmd * 100

print(factory)

from random import shuffle

shuffle(factory)
print(factory)

pack = factory[0:30]
print(len(pack))

unique_colors_in_pack = set(pack)
print(f"Унікальні кольори в пакеті: {unique_colors_in_pack}")
print(f"Кількість унікальних кольорів: {len(unique_colors_in_pack)}")
```

```python
pack = factory[0:30]
counter = 1  # Лічильник для номеру з'їдених M&Ms

while pack:
    print(f"Ми з'їли {counter}-й {pack.pop(-1)} M&Ms")
    counter += 1
    print(f"Залишилося {len(pack)} M&Ms у пакеті")  # Показує кількість залишених цукерок
```

```python
i = 0
pack_number = 1  # Нумерація пачок
while i < len(factory):
    m = []
    for mm_d in factory[i:i+30]:
        m.append(f"Ми з'їли {mm_d} M&Ms")
    print(f"Ми з'їли цілу пачку №{pack_number}! Нам попалось: {m}")
    print(f"У цій пачці було {len(m)} M&Ms\n")  # Додає кількість цукерок у пачці
    i += 30
    pack_number += 1
```
